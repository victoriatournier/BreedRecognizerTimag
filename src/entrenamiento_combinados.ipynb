```python
import pandas as pd

from funciones import *
```

```python
dir_modelos = "modelos/"
dir_csv= "csv/"
```

```python
# Cargar datos de entrenamiento y test
nombre_csv=dir_csv+"gatos_fix.csv"
df = pd.read_csv(nombre_csv)
images_train = df[df['Set'] == 'train']
images_test = df[df['Set'] == 'test']
```

## SIFT

```python
#Parámetros
nombre_csv="sift_"
color=False
textura=False
sift=True
imgz=300
no_clusters=100
```

```python
descriptores_train=calcular_descriptores_train(images_train['Path'], color, textura, sift, imgz, no_clusters, nombre_csv)
df = pd.DataFrame(descriptores_train)
df.to_csv(dir_csv+nombre_csv+"im_features_train.csv", index=False)
descriptores_test=calcular_descriptores_test(images_test['Path'], color, textura, sift, imgz, no_clusters, nombre_csv)
df = pd.DataFrame(descriptores_test)
df.to_csv(dir_csv+nombre_csv+"im_features_test.csv", index=False)
```

```python
descriptores_train=pd.read_csv(dir_csv+nombre_csv+"im_features_train.csv")
descriptores_test=pd.read_csv(dir_csv+nombre_csv+"im_features_test.csv")
# Definir los posibles valores de los hiperparámetros a explorar
param_grid = {
    'C': [1, 0.5, 0.1],  # Parámetro de regularización
    'kernel': ['linear', 'rbf'],  # Tipo de kernel a utilizar
}
svm_model_with_grid_search(
    descriptores_train,
    images_train['Breed'],
    descriptores_test,
    images_test['Breed'],
    param_grid,
    nombre_csv,
)
```
